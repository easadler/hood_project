```python
def fill_elem(name, school):
    rating = df_gs.ix[df_gs['name'] == school , 'gsRating']
    par_rating = df_gs.ix[df_gs['name'] == school , 'parRating']
    df_f.ix[df_f['hood'] == name, 'name'] = school
    df_f.ix[df_f['hood'] == name, 'gsRating'] = int(rating)
    df_f.ix[df_f['hood'] == name, 'parRating'] = int(par_rating)
```

```python
def fill_mid(name, school):
    rating = df_gs.ix[df_gs['name'] == school , 'gsRating']
    par_rating = df_gs.ix[df_gs['name'] == school , 'parRating']
    df_f.ix[df_f['hood'] == name, 'middle'] = school
    df_f.ix[df_f['hood'] == name, 'name_middle'] = school
    df_f.ix[df_f['hood'] == name, 'gsRating_middle'] = int(rating)
    df_f.ix[df_f['hood'] == name, 'parRating_middle'] = int(par_rating)
```

```python
def fill_high(name, school):
    rating = df_gs.ix[df_gs['name'] == school , 'gsRating']
    par_rating = df_gs.ix[df_gs['name'] == school , 'parRating']
    df_f.ix[df_f['hood'] == name, 'high'] = school
    df_f.ix[df_f['hood'] == name, 'name_high'] = school
    df_f.ix[df_f['hood'] == name, 'gsRating_high'] = int(rating)
    df_f.ix[df_f['hood'] == name, 'parRating_high'] = int(par_rating)
```

```python
client = MongoClient()
db = client['neighborhood_recommender']
collection = db['hood_data' + '_' + 'Seattle' + '_' + 'WA']

hood_zillow = [hood for hood in collection.find()]
df_z = pd.DataFrame(hood_zillow)
df_z = df_z.replace('', np.nan)
df_z = df_z.convert_objects(convert_numeric=True)
```

```python
with open('Seattle.json', 'r') as f:
    js = json.load(f)
```

```python
geo_names = []
for geo in js['features']:
    geo_names.append([geo['properties']['NAME'],geo['properties']['REGIONID']])
```

```python
df_g = pd.DataFrame(geo_names, columns =['name', 'id'])
```

```python
x = df_z.ix[df_z['name'].isin(['East Queen Anne','West Queen Anne', 'North Queen Anne']), :].mean()
x['name'] = 'Queen Anne'
x['state'] = 'Washington'
x['city'] = 'Seattle'
x['url'] = 'notexists'
x['_id'] = 99999999999
x['id'] = 271942
df_z = df_z.append(x,ignore_index=True)
```

```python
x = df_z.ix[df_z['name'].isin(['Seaview','Junction']), :].mean()
x['name'] = 'West Seattle'
x['state'] = 'Washington'
x['city'] = 'Seattle'
x['url'] = 'notexists'
x['_id'] = 98989898989
x['id'] = 344020 
df_z = df_z.append(x,ignore_index=True)
```

```python
df_z.ix[df_z['name'] == 'Denny - Blaine', 'name'] = 'Denny-Blaine'
df_z.ix[df_z['name'] == 'Central', 'name'] = 'Central District'
df_z.ix[df_z['name'] == 'Mt. Baker', 'name'] = 'Mount Baker'
```

```python
df_z = df_z.drop(df_z['name'].isin(['Junction','Seaview','Mt. Baker','East Queen Anne','West Queen Anne', 'North Queen Anne']), axis = 0)
```
